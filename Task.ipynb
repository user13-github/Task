```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
boston_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ST0151EN-SkillsNetwork/labs/boston_housing.csv'
df=pd.read_csv(boston_url)
```

```python
df
```

```python
# Assuming 'df' is your dataset
plt.figure(figsize=(6, 8))
sns.boxplot(y=df['MEDV'])
plt.title('Boxplot of Median value for owner-occupied homes (MEDV)')
plt.show()
```

```python
#'CHAS' as the variable related to the Charles river (categorical)
plt.figure(figsize=(6, 4))
sns.countplot(x='CHAS', data=df)
plt.title('Bar plot of Charles river variable')
plt.xlabel('Charles River (Yes/No)')
plt.ylabel('Count')
plt.show()
```

```python
# split the AGE variable into three groups
df['AGE_group'] = pd.cut(df['AGE'], bins=[0, 35, 70, float('inf')], labels=['<=35', ' 35<70', '70>='])

# Boxplot for MEDV vs AGE_group
plt.figure(figsize=(8, 6))
sns.boxplot(x='AGE_group', y='MEDV', data=df)
plt.title('Boxplot of MEDV vs AGE groups')
plt.xlabel('AGE groups')
plt.ylabel('Median value of owner-occupied homes (MEDV)')
plt.show()
```

```python
# 'NOX' for Nitric oxide concentration and 'INDUS' for non-retail business acres
plt.figure(figsize=(8, 6))
sns.scatterplot(x='NOX', y='INDUS', data=df)
plt.title('Scatter plot of Nitric oxide concentration vs Non-retail business acres')
plt.xlabel('Nitric oxide concentration')
plt.ylabel('Proportion of non-retail business acres')
plt.show()
```

```python
# 'PTRATIO' is the variable for pupil to teacher ratio
plt.figure(figsize=(6, 4))
plt.hist(df['PTRATIO'], bins=20, edgecolor='black')
plt.title('Histogram of Pupil to Teacher Ratio')
plt.xlabel('Pupil to Teacher Ratio')
plt.ylabel('Frequency')
plt.show()
```
